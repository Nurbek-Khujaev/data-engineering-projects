```python
class Inventory:
    
    def __init__(self, csv_filename):
        with open(csv_filename, encoding="UTF-8") as f:
            reader = csv.reader(f)
            self.header = next(reader)
            self.rows = list(reader)
            

        price_index = self.header.index("Price")
        for row in self.rows:
            row[price_index] = int(row[price_index])
            
        self.id_to_row = {}
        for row in self.rows:
            self.id_to_row[row[0]] = row
            
        self.prices = set()
        for row in self.rows:
            self.prices.add(row[-1])

    def get_laptop_from_id(self, laptop_id):
        laptop_id = str(laptop_id)  
        for row in self.rows:
            if row[0] == laptop_id:
                return row
        return None
    
    def get_laptop_from_id_fast(self, laptop_id):
        laptop_id = str(laptop_id)
        
        if laptop_id in self.id_to_row:
            return self.id_to_row[laptop_id]
        else:
            return None
        
    def check_promotion_dollars(self, dollars):
        
        price_index = self.header.index("Price")

        for row in self.rows:
            if row[price_index] == dollars:
                return row

        for row1 in self.rows:
            for row2 in self.rows:
                if row1[price_index] + row2[price_index] == dollars:
                    return True

        return False
    
    def check_promotion_dollars_fast(self, dollars):
        if dollars in self.prices:
            return True
        
        for price in self.prices:
            complement = int(dollars) - int(price)
            if complement in self.prices:
                return True
        
        return False 
```

```python
class Inventory:
    
    def __init__(self, csv_filename):
        with open(csv_filename, encoding="UTF-8") as f:
            reader = csv.reader(f)
            self.header = next(reader)
            self.rows = list(reader)
            

        price_index = self.header.index("Price")
        for row in self.rows:
            row[price_index] = int(row[price_index])
            
        self.id_to_row = {}
        for row in self.rows:
            self.id_to_row[row[0]] = row
            
        self.prices = set()
        for row in self.rows:
            self.prices.add(row[-1])
            
        self.rows_by_price = sorted(self.rows, key=lambda row: row[price_index])

    def get_laptop_from_id(self, laptop_id):
        laptop_id = str(laptop_id)  
        for row in self.rows:
            if row[0] == laptop_id:
                return row
        return None
    
    def get_laptop_from_id_fast(self, laptop_id):
        laptop_id = str(laptop_id)
        
        if laptop_id in self.id_to_row:
            return self.id_to_row[laptop_id]
        else:
            return None
        
    def check_promotion_dollars(self, dollars):
        
        price_index = self.header.index("Price")

        for row in self.rows:
            if row[price_index] == dollars:
                return row

        for row1 in self.rows:
            for row2 in self.rows:
                if row1[price_index] + row2[price_index] == dollars:
                    return True

        return False
    
    def check_promotion_dollars_fast(self, dollars):
        if dollars in self.prices:
            return True
        
        for price in self.prices:
            complement = int(dollars) - int(price)
            if complement in self.prices:
                return True
        
        return False
    
    def find_first_laptop_more_expensive(self, target_price):
        range_start = 0
        range_end = len(self.rows_by_price) - 1
        result_index = -1  

        while range_start <= range_end:
            range_middle = (range_start + range_end) // 2
            middle_price = self.rows_by_price[range_middle][-1]

            if middle_price > target_price:
                result_index = range_middle
                range_end = range_middle - 1  
            else:
                range_start = range_middle + 1

        return result_index
        
```
